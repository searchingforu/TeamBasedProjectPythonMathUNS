### Cluster

```python
ds = pd.read_csv('Kasus_DBD_Indonesia_2004_2023.csv', index_col ="Provinsi" )
ds
```

```python
x =ds.iloc[:, [0,1,2,3,4]].values
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++',max_iter=200,n_init=10,random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11), wcss)
plt.title("Elbow Method")
plt.ylabel("WCSS")
plt.xlabel("Number of Clusters")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init = "k-means++", max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100 ,c ='black', label= 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100 ,c ='blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100 ,c ='grey', label = 'Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 100 c='yellow', label = 'Centroids')
plt.legend()
```

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s=100, c='black', label='Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s=100, c='blue', label='Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s=100, c='grey', label='Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='yellow', label='Centroids')
plt.legend()
```

```python
ds = pd.read_csv('Kasus_DBD_Indonesia_2004_2023.csv', index_col ="Provinsi" )
ds
```

```python
years = ["2019", "2020", "2021", "2022", "2023"]

# Loop through the years to plot each pair
for i in range(len(years) - 1):
    plt.scatter(ds[years[i]], ds[years[i + 1]], s=50, label=f"{years[i]} vs {years[i + 1]}")

plt.xlabel("Year 1")
plt.ylabel("Year 2")
plt.title("Kasus DBD")
plt.legend()  # Add a legend to distinguish the plots
plt.show()
```

```python
from itertools import combinations

for year1, year2 in combinations(years, 2):
    plt.scatter(ds[year1], ds[year2], s=50, label=f"{year1} vs {year2}")
plt.xlabel("2020")
plt.ylabel("2019")
plt.title("Kasus DBD")
plt.legend()
plt.show()
```
